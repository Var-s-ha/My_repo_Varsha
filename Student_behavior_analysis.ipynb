### Metrics:

* Davies-Bouldin index (DBI)
An internal evaluation metric that measures the separation between clusters and the tightness of the clusters. A lower value indicates tighter clusters and better separation. 
* Calinski-Harabasz index
A metric that measures the quality of clustering results by comparing the ratio of inter-cluster variances to intra-cluster variances. It also includes a penalty factor for the number of clusters to avoid overfitting. 
* Silhouette coefficient
A metric that measures the performance of clustering by calculating the mean intra-cluster distance and the mean inter-cluster distance to the closest cluster.

```python
opt_cluster_data = final_data.copy()
```

```python
opt_cluster_data = opt_cluster_data.drop(columns=['cluster'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.metrics import pairwise_distances

# Perform hierarchical clustering
Z = linkage(opt_cluster_data, method='ward')  # 'ward' minimizes the variance within clusters

def compute_wcss(X, labels):
    """Compute the within-cluster sum of squares (WCSS) for given labels."""
    wcss = 0
    for label in np.unique(labels):
        cluster_points = X[labels == label]
        centroid = np.mean(cluster_points, axis=0)
        wcss += np.sum((cluster_points - centroid) ** 2)
    return wcss

# Evaluate WCSS for a range of cluster numbers
max_clusters = 10
wcss_values = []

for n_clusters in range(2, max_clusters + 1):
    labels = fcluster(Z, n_clusters, criterion='maxclust')
    wcss = compute_wcss(opt_cluster_data.values, labels)  # df is your data
    wcss_values.append(wcss)

# Plotting the Elbow Plot
plt.figure(figsize=(8, 5))
plt.plot(range(2, max_clusters + 1), wcss_values, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Clusters in Hierarchical Clustering')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```
